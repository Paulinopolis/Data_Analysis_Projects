```python
#Création d'une nouvelle table transpose
```

```python
cm_transpose = citymapper.transpose()
```

```python
cm_transpose
```

```python
cm_transpose.index
```

```python
new_header = cm_transpose.iloc[0] #grab the first row for the header
cm_transpose = cm_transpose[1:] #take the data less the header row
cm_transpose.columns = new_header #set the header row as the df header
```

```python
#Creation of the Column Country
```

```python
#Creation of the values for Country and Continent
country  = ['Netherlands','Spain','Germany', 'England','USA','Belgium', 'USA', 'Denmark', 'Germany', 
            'China', 'Turkey','Portugal','England','USA', 'France', 'Spain' , 'England','Australia', 'Mexico', 
            'Italia', 'Monaco', 'Canada', 'Russia','USA', 'France', 'USA', 'Germany', 'Italy','USA', 
            'USA', 'South Korea', 'Singapore', 'Russia','Sweden', 'Australia', 'Brezil', 'Japan', 'Canada','Canada',
            'Austria', 'USA']

continent  = ['Europe','Europe','Europe', 'Europe','America','Europe', 'America', 'Europe', 'Europe', 
            'Asia', 'Europe','Europe','Europe','America', 'Europe', 'Europe' , 'Europe','Europe', 'America', 
            'Europe', 'Europe', 'America', 'Europe','America', 'Europe', 'America', 'Europe', 'Europe','America', 
            'America', 'Asia', 'Asia', 'Europe','Europe', 'Asia', 'America', 'Asia', 'America','America',
            'Europe', 'America']
```

```python
cm_transpose2 = cm_transpose
```

```python
#Column Title transformation into date format
cm_transpose2.columns = pd.to_datetime(cm_transpose2.columns)
```

```python
#Creation of the Column Country
cm_transpose2.insert(0,"country",country, allow_duplicates=False)
```

```python
#Creation of the Column Continent
cm_transpose2.insert(1,"continent",continent, allow_duplicates=False)
```

```python
cm_transpose2
```

```python
#Modification of the datatype for the columns with date titles. 
for i in [i for i in list(range(len(cm_transpose2.columns))) if i not in [0,1]]:
    cm_transpose2.iloc[:,i]=pd.to_numeric(cm_transpose2.iloc[:,i])
```
