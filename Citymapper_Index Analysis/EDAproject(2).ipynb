```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_viz = pd.read_csv('citymapperproject.csv')
```

```python
#renaming the first column
data_viz.rename(columns = {'Unnamed: 0':'cities'}, inplace = True)
```

```python
#selecting data for march (02/03 to 31/03)
data_viz2 = data_viz[['cities', 'country', 'continent', '2020-03-02 00:00:00',
       '2020-03-03 00:00:00', '2020-03-04 00:00:00', '2020-03-05 00:00:00',
       '2020-03-06 00:00:00', '2020-03-07 00:00:00', '2020-03-08 00:00:00',
       '2020-03-09 00:00:00', '2020-03-10 00:00:00', '2020-03-11 00:00:00',
       '2020-03-12 00:00:00', '2020-03-13 00:00:00', '2020-03-14 00:00:00',
       '2020-03-15 00:00:00', '2020-03-16 00:00:00', '2020-03-17 00:00:00',
       '2020-03-18 00:00:00', '2020-03-19 00:00:00', '2020-03-20 00:00:00',
       '2020-03-21 00:00:00', '2020-03-22 00:00:00', '2020-03-23 00:00:00',
       '2020-03-24 00:00:00', '2020-03-25 00:00:00', '2020-03-26 00:00:00',
       '2020-03-27 00:00:00', '2020-03-28 00:00:00', '2020-03-29 00:00:00',
       '2020-03-30 00:00:00', '2020-03-31 00:00:00', 'StartLockdown', 'LType']]
```

```python
#selecting data of 7 cities/country/continent (except Africa)
data_viz3 = data_viz2.loc[data_viz2['cities'].isin(['Barcelona', 'Mexico City','Sydney','Tokyo','Seattle' ,'Moscow', 'Singapore'])]
```

```python
data_viz3.describe()
```

```python
#adding column 'sum'
data_viz4 = data_viz3.assign(Total_index=data_viz3.sum(axis=1,numeric_only=True))
```

```python
data_viz4.head()
```

```python
data_viz5 = data_viz4[['cities', 'LType', 'Total_index']]
```

```python
data_viz5
```

```python
#2) histogramme
sns.histplot(x='cities', y='Total_index', data=data_viz5)
plt.xticks(rotation=45)
plt.title ('Histogram of Total Index/cities in March 2020')
```

```python
#4) barplot
sns.barplot(x='citie', y='Total_index', data=data_viz5)
plt.title ('Barplot of Total Index/cities in March 2020')
plt.xticks(rotation=45)
```

```python
#1) scatter plot
sns.scatterplot(data = data_viz5, x='cities', y='Total_index')
plt.show()
```

```python
#3)box plot
sns.boxplot(data = data_viz3)
plt.xticks(rotation=90)
plt.title ('Boxplot of the 7 cities indexes per date')
```

```python
#hist lockdown
sns.histplot(x='LType', y='Total_index', data=data_viz5)
plt.xticks(rotation=45)
plt.title ('Histogram of Total Index/Lockdowntype in March 2020')
```

```python
#scatterplot lockdown
sns.scatterplot(data = data_viz5, x='LType', y='Total_index')
plt.title ('Scatterplot of Total Index/Lockdowntype in March 2020')
plt.show()
```
