```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
chrome_options = webdriver.ChromeOptions()
#chrome_options.add_argument("--headless")
# below trick saved my life
chrome_options.add_argument('--no-sandbox')

chrome_options.add_argument("start-maximized")

# Optional argument, if not specified will search path.
driver = webdriver.Chrome(executable_path='/Users/paulinopolis/Documents/Week9/chromedriver',
                          options=chrome_options)
```

```python
import time
```

```python
#RECUPERER LES ELEMENTS DES PAGES
```

```python
#RECUPERER LE COMMENTAIRE
```

```python
com_amz = []
page = 1
while page < 5:
    driver.get(f"https://www.amazon.fr/France-pas-dit-son-dernier/product-reviews/2957930501?pageNumber={str(page)}")
    time.sleep(20)
    comments = driver.find_element_by_xpath("/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]")
    time.sleep(10)
    for c in comments.find_elements_by_css_selector('span.review-text-content'):
        com_amz.append(c.text)
        print(c.text)
    page += 1 
```

```python
#RECUPERER LA DATE
```

```python
dates_amz = []
page = 1
while page < 5:
    driver.get(f"https://www.amazon.fr/France-pas-dit-son-dernier/product-reviews/2957930501?pageNumber={str(page)}")
    time.sleep(20)
    dates = driver.find_element_by_xpath("/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]")
    time.sleep(10)
    for d in dates.find_elements_by_css_selector('span.review-date'):
        dates_amz.append(d.text)
        print(d.text)
    page += 1   
```

```python
#RECUPERER LES LIKES
```

```python
likes_amz = []
page = 1
while page < 5:
    driver.get(f"https://www.amazon.fr/France-pas-dit-son-dernier/product-reviews/2957930501?pageNumber={str(page)}")
    time.sleep(20)
    likes = driver.find_element_by_xpath("/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]")
    time.sleep(10)
    for l in likes.find_elements_by_css_selector('span.cr-vote-text'):
        likes_amz.append(l.text)
        print(l.text)
    page += 1 
```

```python
#RECUPERER LE TITRE DU COMMENTAIRE
```

```python
title_amz = []
page = 1
while page < 5:
    driver.get(f"https://www.amazon.fr/France-pas-dit-son-dernier/product-reviews/2957930501?pageNumber={str(page)}")
    time.sleep(20)
    title = driver.find_element_by_xpath("/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]")
    time.sleep(10)
    for t in title.find_elements_by_css_selector('a.review-title-content'):
        title_amz.append(t.text)
        print(t.text)
    page += 1 
```

```python
#CREER LE PREMIER DATAFRAME
```
